```python

# List of parquet file types to read
parquet_files = {
    "games": "games/games_20211019_20250924_021536.parquet",
    "players": "players/players_20250924_023542_current.parquet",
    "player_logs": "player_logs/player_logs_20211019_20250924_024839_game_20211019_BKN@MIL.parquet",
    "teams": "teams/teams_20250924_021516_current.parquet"
}

# Read each parquet file into a DataFrame and store in a dictionary
dfs = {}
for key, rel_path in parquet_files.items():
    full_path = data_path + rel_path
    try:
        dfs[key] = pd.read_parquet(full_path)
        print(f"{key} DataFrame loaded. Shape: {dfs[key].shape}")
    except Exception as e:
        print(f"Error loading {key} from {full_path}: {e}")

# Optionally, display the first few rows of each DataFrame
for key, df in dfs.items():
    print(f"\n--- {key.upper()} ---")
    display(df.head())

```
